```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
from adjustText import adjust_text
import matplotlib.ticker as mticker
import seaborn as sns
#load the data
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
dfj = dataset['train'].to_pandas()
#data clean-up
dfj['job_posted_date']=pd.to_datetime(dfj['job_posted_date'])
dfj['job_skills'] = dfj['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

dfj_DA = dfj[dfj['job_title_short']=='Data Analyst'].copy() # calling specificallly the role from the column for the remainder of the lessons
```

```python
dfj_DA_US = dfj[(dfj['job_title'] == 'Data Analyst') & (dfj['job_country'] == 'United States')].copy()
```

```python
dfj_DA_US['job_posted_month_no'] = dfj_DA_US['job_posted_date'].dt.month

# Explode job_skills column
dfj_DA_US_explode = dfj_DA_US.explode('job_skills')
```

```python
dfj_DA_US_pivot = dfj_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills',  aggfunc='size', fill_value=0)
dfj_DA_US_pivot.loc['Total'] = dfj_DA_US_pivot.sum()
dfj_DA_US_pivot = dfj_DA_US_pivot[dfj_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]
dfj_DA_US_pivot = dfj_DA_US_pivot.drop('Total')

dfj_DA_US_pivot

```

```python
dfj_DA_Totals = dfj_DA_US.groupby('job_posted_month_no').size()
dfj_DA_Totals
```

```python
dfj_DA_US_percent = dfj_DA_US_pivot.div(dfj_DA_Totals/100, axis=0)
dfj_DA_US_percent
```

```python
dfj_DA_US_percent = dfj_DA_US_percent.reset_index()
dfj_DA_US_percent['job_posted_month'] = dfj_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
dfj_DA_US_percent = dfj_DA_US_percent.set_index('job_posted_month')
dfj_DA_US_percent = dfj_DA_US_percent.drop(columns='job_posted_month_no')

dfj_DA_US_percent
```

```python
from matplotlib.ticker import PercentFormatter

dfj_plot = dfj_DA_US_percent.iloc[:, :5]
sns.lineplot(data=dfj_plot, dashes=False, legend='full', palette='tab10')
sns.set_theme(style='ticks')
sns.despine() # remove top and right spines

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# annotate the plot with the top 5 skills using plt.text()
for i in range(5):
    plt.text(11.2, dfj_plot.iloc[-1, i], dfj_plot.columns[i], color='black')

plt.show()
```
